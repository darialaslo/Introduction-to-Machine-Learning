```python
import scipy as sp
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data_train = pd.read_csv("./train.csv")
```

```python
data_train.iloc[:,1]
```

```python
feature_names = []
for i in range(21):
    name = "x" + str(i+1)
    feature_names.append(name)
    
def featurize(data):
    final_features = pd.DataFrame(index=range(data.shape[0]), columns=feature_names)
    features = data.iloc[:, 2:7]
    final_features.iloc[:, 0:5] = features
    final_features.iloc[:, 5:10] = np.power(features, 2).values
    final_features.iloc[:, 10:15] = np.exp(features).values
    final_features.iloc[:, 15:20] = np.cos(features).values
    final_features.iloc[:, -1] = 1
    return final_features
    
featurize(data_train)
```

```python
linreg = LinearRegression().fit(featurize(data_train), data_train['y'])
submission = pd.DataFrame(linreg.coef_)
submission.to_csv('./submission.csv', index=False, header=False)
```
